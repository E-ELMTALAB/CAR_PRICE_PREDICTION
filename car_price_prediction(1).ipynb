```python
df_droped = df.drop(columns = ["Model" , "Fuel Type" , "Location" , "Color" , "Seller Type" , "Make"] , index=1)
```

```python
import re

def extract_number(value):

  # print(value)
  pattern = r"(\d+)"
  if isinstance(value , str):
    match = re.search(pattern, value)
    if match:
        number_before_bhp = int(match.group(1))
        return number_before_bhp

df_droped["Max Power"] = df_droped["Max Power"].apply(extract_number)
df_droped["Max Torque"] = df_droped["Max Torque"].apply(extract_number)
df_droped["Engine"] = df_droped["Engine"].apply(extract_number)
```

```python
df_droped["Drivetrain"].value_counts()
```

```python
df_droped["Transmission"] = df_droped["Transmission"].replace(["Automatic" , "Manual"] , [0 , 1])
df_droped["Owner"] = df_droped["Owner"].replace(["First" , "Second" , "Third" , "UnRegistered Car" , "Fourth" , "4 or More"] , [0 , 1 , 2, 3, 4 , 5])
df_droped["Drivetrain"] = df_droped["Drivetrain"].replace(["FWD" , "RWD" , "AWD"] , [0 , 1 , 2])
```

```python
df_droped.isna().sum()
```

```python
# List of columns for which you want to fill missing values with the mode
cat_columns_to_fill = ['Drivetrain', 'Seating Capacity']

# Fill missing values with the mode for the specified columns
for column in cat_columns_to_fill:
    df_droped[column] = df_droped[column].fillna(df_droped[column].mode().iloc[0])

num_columns_to_fill = ['Engine', 'Max Torque' , 'Max Power' , 'Length' , 'Width' , 'Height' , 'Fuel Tank Capacity']

# Fill missing values with the mode for the specified columns
for column in num_columns_to_fill:
    df_droped[column] = df_droped[column].fillna(df_droped[column].mean())
```

```python
df_droped.to_csv("/content/output.csv" , index = False)
```

```python
cat_columns = ["Transmission" , "Owner" , "Drivetrain" , "Seating Capacity"]

df_one_hot = pd.get_dummies(df_droped , columns = cat_columns)
```

```python
num_columns = df_one_hot.select_dtypes(["int" , "float"]).columns.values.tolist()
num_columns.remove("Price")
```

```python
num_columns
```

```python
scaler = StandardScaler()

df_one_hot[num_columns] = scaler.fit_transform(df_one_hot[num_columns])
df_scaled = df_one_hot
```

```python
df_scaled.head(10)
```

```python
df_scaled["Price"] = df_scaled["Price"].apply(lambda x : x/1000000)
```

```python
train_columns = df_scaled.columns.values.tolist()
train_columns.remove("Price")
# print(columns)
X = df_scaled[train_columns]
y = df_scaled["Price"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
y_train
```

```python
from tensorflow.keras.metrics import MeanSquaredError, MeanAbsoluteError, RootMeanSquaredError

# Step 4: Build the TensorFlow Regression Model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(128, activation='relu', input_shape=(X_train.shape[1],)),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(1)  # Single output neuron for regression
])

# Step 5: Compile the Model
model.compile(optimizer='adam', loss='mean_squared_error' , metrics=[MeanSquaredError() , MeanAbsoluteError() , RootMeanSquaredError()])

# Step 6: Train the Model
num_epochs = 100
batch_size = 32
model.fit(X_train, y_train, epochs=num_epochs, batch_size=batch_size, validation_data=(X_test, y_test))
```

```python
# Step 7: Evaluate the Model
loss = model.evaluate(X_test, y_test)
print(f"Mean Squared Error: {loss}")
```
